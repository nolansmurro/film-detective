```python
import os
```

```python
def print_file_tree(startpath, max_depth=3):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        if level > max_depth:  # Limit the depth of the traversal
            continue
        indent = ' ' * 4 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 4 * (level + 1)
        for f in files:
            print(f"{subindent}{f}")
```

```python
ROOT_DIR = root_dir = os.path.dirname(os.path.dirname(os.getcwd()))

MAX_DEPTH = 3
print_file_tree(ROOT_DIR, MAX_DEPTH)
```
